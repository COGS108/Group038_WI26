## Research Question

How did the popularity of each different video game mode (singleplayer, multiplayer, online co-op) on PC change between the pre-COVID period (2018-2019), the COVID period (2020-2021), and post-COVID period (2022-2023) among the top player-count Steam games given 50 games from each category?

where we are defining **popularity** by metrics of:
- Average concurrent player count over a given period 
- Peak player count over a given period
- Average play time per player per week or period
- Achievement completion rate
- Number of new players during each period
- Search interest in a game (e.g. Google Trends)

## Background and Prior Work

The global coronavirus outbreak in 2020, called COVID-19, has caused a global pandemic, forcing people to isolate and quarantine from each other. During the lockdown period, people spent most of their time at home and turned to digital entertainment and video games as a way to socialize and de-stress. Therefore, the gaming industry in this period witnessed a peak in gamer activity, play time, sales, and stock values.

In this study, we are interested in finding which games were the most popular from each category, seeing what game type people are interested in the most, which could give the gaming industry a better understanding of their consumer's desires. At first, the group was interested in researching the top 50 games from each category, between the pre-COVID period (2018-2019), the COVID period (2020-2021), and the post-COVID period (2021-2023), across multiple popular platforms such as Steam, Epic Games, Xbox, PlayStation, Nintendo, etc. **But** not all platforms share their players' statistics to the public, so we narrowed down to only Steam, which has a public database called "Steam Charts" and "SteamDB".

There are multiple research papers conducted about games' activity and price range analysis over the course of time, such as: work from Aliev, Aliyev, and Eyniyev. However, they mainly focus on the Indie games category, while treating AAA - a pricing category - as a game category. A paper from Sener, Yalcin, and Gulseven successfully showed a significant rise in Steam activies during the COVID-19 era, but did not close in on any other said parameters. Neither of the papers mentioned are hitting the main points of analyzing the top popular games from each category.

Therefore, as gamers, we decided to conduct extensive research on what are the top 50 popular games from each category on Steam pre/during/post COVID-19 eras.
